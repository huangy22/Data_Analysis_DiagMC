#### Two-Spin System

```python
def two_spin_HTE(x, order):
    
    terms = [x/4, - x**2/16, - x**3/64, x**4/768, (5*x**5)/3072, 
             (13*x**6)/61440, -(77*x**7)/737280 , -(823*x**8)/20643840, (25*x**9)/16515072 ,
             (11593*x**10)/2972712960, (37873*x**11)/59454259200, -(590473*x**12)/2615987404800,
             -(91741*x**13)/896909967360, (143321*x**14)/1632376140595200,(422870171*x**15)/45706531936665600,
            (5011276729*x**16)/2742391916199936000, -(94294597*x**17)/199446684814540800,
             -(14847332219*x**18)/57379277015875584000, -(4738273523*x**19)/548030645784281088000,
            (44176120892627*x**20)/2040866124900662771712000]
    return np.sum(3*terms[0:order])
```

```python
def exact_two_spins(beta):
    return 3*beta/(np.exp(beta)+3.0)
```

```python
DiagMC_Order = np.array([1, 3, 4, 5, 6])
DiagMC = np.array([0.2688*3, 0.755, 0.7077, 0.6899, 0.6687])
DiagMC_err = np.array([0.0006, 0.002, 0.0002, 0.0003, 0.0002])
```

```python
T_list = np.linspace(0.3, 2.0, 100)
HTE_list = np.zeros((5, len(T_list)))
```

```python
for t in range(len(T_list)):
    for i in range(5):
        order = i+10
        HTE_list[i,t] = two_spin_HTE(1.0/T_list[t], order)
```

```python
plt.figure()
for i in range(5):
    plt.plot(T_list, HTE_list[i], label="Order "+str(i+10))
#y=[0.0937,]*MaxTauBin
#plt.plot(TGrid, y, 'k--', label="Unbiased MC")
plt.ylim([0.0,0.9])
plt.xlim([0,2.2])
plt.xlabel("T/J")
plt.ylabel(r"$\chi(k=0, i\omega_n=0)$")
plt.legend()
plt.title("High-Temperature Expansion")

plt.savefig("High-Temperature.pdf")
plt.show()
```

```python
HTE_Order = np.array(range(1, 15))
HTE = np.array([two_spin_HTE(2.0, order) for order in HTE_Order])
```
